# **Neural Networks From Scratch**

*demonstrates forward and backward propagation without using any external libraries like tensorflow or pytorch*

*in this example i have considered XOR gate to demonstrate how nural network can learn complex decision boundaries*

```python
import numpy as np
x= np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y = np.array([
    [0],
    [1],
    [1],
    [0]])
y.shape,x.shape
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def derivative_sigmoid(z):
    return sigmoid(z)*(1-sigmoid(z))
```

```python
weights =2* np.random.random((2,1))-1
biase = 2* np.random.random(1)-1
out=x
output = sigmoid(np.dot(out,weights)+biase)
lr=0.1
weights,biase,output
```

```python
# gradient decent
for i in range(10000):
    first_term = output-y
    in_last_layer = np.dot(out,weights)+biase
    second_term = derivative_sigmoid(in_last_layer)
    first_two = first_term*second_term


    changes = np.array([[0.0],[0.0]])
    for i in range(2):
        for j in range(4):
            changes[i][0] += first_two[j][0] * out[j][i]
    weights= weights-lr*changes

    bias_change=0.0
    for i in range(2):
        for j in range(4):
            changes[i][0] += first_two[j][0]
    biase=biase-lr*bias_change
output=sigmoid(np.dot(x,weights)+biase)
weights,biase,output
```

```python
# gradient decent optimised

for i in range(10000):
    first_term = output-y
    in_last_layer = np.dot(out,weights)+biase
    second_term = derivative_sigmoid(in_last_layer)
    first_two = first_term*second_term


    changes = np.dot(out.T,first_two)

    bias_change= np.sum(first_two)
    biase=biase-lr*bias_change

output=sigmoid(np.dot(x,weights)+biase)
weights,biase,output
```

```python
wh = 2* np.random.random((2,2))-1
bh = 2* np.random.random((1,2))-1

wo = 2* np.random.random((2,1))-1
bo = 2* np.random.random((1,1))-1
```

```python
for i in range(1000):
    out=x
    input_hidden = np.dot(out,wh) + bh
    output_hidden = sigmoid(input_hidden)
    input_for_output_layer=np.dot(output_hidden,wo) + bo
    output = sigmoid(input_for_output_layer)

    first_term_output_layer=output-y
    second_term_output_layer= derivative_sigmoid(input_for_output_layer)
    first_two_output_layer=first_term_output_layer*second_term_output_layer

    first_term_hidden_layer = np.dot(first_two_output_layer,wo.T)
    second_term_hidden_layer = derivative_sigmoid(input_hidden)
    first_two_hidden_layer = first_term*second_term


    changes_output = np.dot(output_hidden.T,first_two_output_layer)
    changes_output_bias = np.sum(first_two_output_layer,axis=0 ,keepdims=True)

    changes_hidden = np.dot(out.T,first_two_hidden_layer)
    changes_hidden_bias = np.sum(first_two_hidden_layer,axis=0 ,keepdims=True)

    wo = wo - lr* changes_output
    bo = bo - lr* changes_output_bias

    wh = wo - lr* changes_output
    bh = bh - lr* changes_hidden_bias

out=x
input_hidden = np.dot(out,wh) + bh
output_hidden = sigmoid(input_hidden)
input_for_output_layer=np.dot(output_hidden,wo) + bo
output = sigmoid(input_for_output_layer)

output,wh,bh,wo,bo
```
